```python
# 1. Hapus data
final_normal = final_normal.drop(index=normal_deleted)
final_abnormal = final_abnormal.drop(index=abnormal_deleted)

# 2. Pindahkan data dari normal ke abnormal
move_normal_to_abnormal = final_normal.loc[normal_to_abnormal]
final_abnormal = pd.concat([final_abnormal, move_normal_to_abnormal])
final_normal = final_normal.drop(index=normal_to_abnormal)

# 4. Reset index (opsional supaya rapi)
final_normal = final_normal.reset_index(drop=True)
final_abnormal = final_abnormal.reset_index(drop=True)
```

```python
def get_image_url(image_id):
  return f'https://storage-api.its.ac.id/public/22002468891c4b189caf102b20a1d72c/{image_id}/image'

def download_image(image_id, idx, kondisi):
  url = get_image_url(image_id)
  response = requests.get(url)
  img = Image.open(BytesIO(response.content))
  img.save(f'/content/drive/MyDrive/dataset/images/images_v2/{kondisi}/{kondisi}-{idx}.jpg')
```

```python
for idx, row in final_normal.iterrows():
  download_image(row['foto_cairan'], idx, 'normal')
```

```python
for idx, row in final_abnormal.iterrows():
  download_image(row['foto_cairan'], idx, 'abnormal')
```
